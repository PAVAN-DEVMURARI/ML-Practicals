```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

**Apply Feature Construction**

```python
X['Family_size'] = X['SibSp'] + X['Parch'] +1
```

```python
X.sample(10)
```

```python
def myfunc (num):
  if num == 1:
    return 0
  elif num > 1 and num <= 4:
    return 1
  else:
    return 2
```

```python
myfunc(2)
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.sample(10)
```

```python
X.drop(columns = ['SibSp','Parch','Family_size'],inplace=True)
```

```python
X.sample(10)
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=210))
```
