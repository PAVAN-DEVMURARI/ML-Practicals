```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/SMSSpamCollection', sep="\t", header=None, names=["label", "message"])
```

```python
df.info()
```

```python
df.sample(10)
```

```python
df['label'].value_counts()
```

```python
df['message'] = df['message'].str.lower()
```

```python
df["tokens"] = df["message"].apply(lambda x: x.split())
```

```python
df.sample(10)
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Remove stop words from tokenized column
df["tokens"] = df["tokens"].apply(lambda tokens: [t for t in tokens if t.lower() not in ENGLISH_STOP_WORDS])
```

```python
df.sample(10)
```

```python
df.drop(columns=["message"], inplace=True)
```

```python
df.sample(10)
```

```python
df['label'] = df['label'].map({'spam': 1, 'ham': 0})
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def identity_tokenizer(tokens):
    return tokens


vectorizer = CountVectorizer(tokenizer=identity_tokenizer, lowercase=False)

X_counts = vectorizer.fit_transform(df["tokens"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X_counts , df['label'] , test_size=0.2 , random_state=42, stratify=df['label'])
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svm_clf = SVC(kernel='linear', C=1.0)
```

```python
svm_clf.fit(X_train, y_train)
```

```python
y_pred = svm_clf.predict(X_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

acc = accuracy_score(y_test, y_pred)

prec = precision_score(y_test, y_pred)

rec = recall_score(y_test, y_pred)

f1 = f1_score(y_test, y_pred)

print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1 Score : {f1:.4f}")
```
